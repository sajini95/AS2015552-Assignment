```python
#load data set and handle missing values if any
from scipy.io import arff
import pandas as pd
import numpy as np

data = arff.loadarff('EEG_Eye_State.arff')
eyeData = pd.DataFrame(data[0])

eyeData = eyeData.replace('[?]', np.nan , regex = True)
eyeData = eyeData.dropna()
eyeData
```

```python
dataArray = np.array(eyeData)
```

```python
data = dataArray[:,0:14]
data
```

```python
target = dataArray[:,14]
target
```

```python
data.shape
```

```python
reshapedData = data.reshape(14980,14,1)
reshapedData.shape
```

```python
target.shape
```

```python
#split data into train and test
split1 = 0.8 * len(dataArray)
split1 = int(split1)

xtrain = reshapedData[0:split1,0:14]
ytrain = target[0:split1]
xtest = reshapedData[split1 + 1:len(dataArray),0:14]
ytest = target[split1 + 1:len(dataArray)]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

#model = Sequential()
#model.add(LSTM((1),batch_input_shape = (14980,14,1), return_sequences = False))
#model.compile(loss = 'mean_absolute_error', optimizer = 'adam', metrics = ['accuracy'])
#model1 = model.fit(xtrain,ytrain,epochs = 300 )
```

```python
#model.summary()
```

```python
model1 = Sequential()
model1.add(LSTM(1, input_shape=(14,1)))
model1.add(Dense(1, activation='linear'))
# compile model
model1.compile(loss ='mean_absolute_error', optimizer='adam', metrics = ['accuracy'])
```

```python
model1.fit(xtrain, ytrain, epochs = 50, shuffle = False, verbose=1)
```

```python
results = model1.predict(xtest)
results.astype(int)
```
